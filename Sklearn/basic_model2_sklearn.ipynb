```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.DataFrame({
    'Edad': [25, 30, 35, 40, 45, 50, 55, 60],
    'Sexo': [0, 1, 0, 1, 0, 1, 0, 1],
    'Educación': [2, 3, 4, 5, 6, 7, 8, 9],
    'Salario': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000]
})
```

```python
#Division del conjunto de datos en caracteristicas (X) e (Y)
x = datos[['Edad', 'Sexo', 'Educación']]
y = datos['Salario']
```

```python
#Division entre conjunto de datos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Modelo de regresion lineal multiple
modelo = LinearRegression()

#Entrenamiento del modelo
modelo.fit(x_train, y_train)
```

```python
#Predicion del salario
salario_predicho = modelo.predict(x_test)
print(salario_predicho)
```

```python
#Evaluacion del modelo
print(f"Coeficiente de determinación (R^2): {modelo.score(x_test, y_test):.2f}")
```

```python
sns.set()
sns.scatterplot(x='Edad', y='Salario', data=datos)
plt.plot(x_test['Edad'], salario_predicho, color='Green')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Establece el estilo de la gráfica
sns.set_theme(style="dark")

# Genera datos aleatorios
np.random.seed(0)
n = 100
x = np.random.rand(n)
y = 3 * x + 2 + np.random.randn(n) / 1.5

# Divide los datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Crea y entrena el modelo
modelo = LinearRegression()
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
modelo.fit(x_train, y_train)

# Predice valores
y_pred = modelo.predict(x_test)

# Crea la figura y el eje
fig, ax = plt.subplots(figsize=(8, 8))

# Establece el fondo como una galaxia
ax.set_facecolor((0.1, 0.1, 0.3))  # Color azul oscuro

# Grafica los datos reales como estrellas
sns.scatterplot(x=x_test.flatten(), y=y_test, s=5, color="yellow", label='Datos reales', alpha=0.8)

# Grafica las predicciones como planetas
sns.scatterplot(x=x_test.flatten(), y=y_pred, s=5, color="blue", label='Predicciones', alpha=0.8)

# Grafica el agujero negro
circle = plt.Circle((0.5, 2), 0.2, edgecolor='black', facecolor='black', alpha=0.8)
ax.add_artist(circle)

# Grafica la línea de regresión como una trayectoria de entrada al agujero negro
sns.regplot(x=x_test.flatten(), y=y_test, color="red", line_kws={'linewidth': 1, 'linestyle':'--'})

# Grafica la densidad de los datos como una nebulosa
sns.kdeplot(x=x_test.flatten(), y=y_test, levels=5, color="purple", linewidths=1, alpha=0.5)

# Establece los límites del eje
ax.set_xlim(0, 1)
ax.set_ylim(0, 4)

ax.set_title('data black hole')
ax.set_xlabel('Variable Independiente')
ax.set_ylabel('Variable Dependiente')

plt.legend()
plt.show()
```
