```python
#Importacion de librerias
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Carga del conjunto de datos
datos = np.array([
    [100, 200000],
    [120, 240000],
    [150, 300000],
    [180, 360000],
    [200, 400000]
])
print(datos)
```

```python
#Division del conjunto de datos en caracteristicas (x) y Etiquetas (y)
x = datos[:,0].reshape(-1,1)
y = datos[:,1]
print(f'X:{x}\nY:{y}')
```

```python
#Division del conjunto de datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print(f'Xtrain:\n{x_train}\nXtest:{x_test}\nYtrain:{y_train}\nYtest:{y_test}')
```

```python
#Creacion del modelo de regresion lineal
modelo = LinearRegression()
#Entrenamiento del modelo
modelo.fit(x_train, y_train)
```

```python
#Prediccion del precio para un tamaño nuevo
tamaño_nuevo = 160
precio_predicho = modelo.predict(np.array([[tamaño_nuevo]]))
print(f'Precio predicho para un Tamaño de {tamaño_nuevo} m2:{precio_predicho[0]:.2f} USD')
```

```python
#Visualizacion del modelo
plt.scatter(x, y)
plt.plot(x, modelo.predict(x), color='red')
plt.xlabel('Tamaño (m2)')
plt.ylabel('Precio (USD)')
plt.show()
```
