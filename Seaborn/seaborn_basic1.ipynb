COMANDOS BASICOS SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Graficar datos simples:

Grafico de dispersion (scatter plot)

```python
sns.scatterplot(x=[1,2,3,4], y=[10,20,25,30])
plt.show()
```

Grafico de barras

```python
sns.barplot(x=['A','B','C'], y=[5,10,15])
plt.show()
```

Grafico de caja (box plot)

```python
sns.boxenplot(x=[1,2,3,4,5], y=[10,20,25,30,35])
plt.show()
```

Histograma (distplot)

```python
import numpy as np
data = np.random.normal(size=100)
sns.histplot(data, kde=True)#Histograma con curva de densidad (KDE)
plt.show()
```

Graficos Relacionales (Replot)

```python
#Dataset 'tips' de ejemplo que trae Seaborn
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips)
plt.show()
```

Grafico de regresion (regplot)

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

Graficar Datos Categoricos:

Grafico de violin (violin plot)

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

Graficos de puntos (strip plot)

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

Mapa de calor (heatmap)

```python
data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, cmap='YlGnBu')#Mapa de calor con anotaciones
plt.show()
```

Pares de graficos (pairplot)

```python
sns.pairplot(tips)
plt.show()
```

Ajustes Adicionales

Cambiar el estilo del grafico

```python
sns.set_style("whitegrid")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

Añadir Titulos y Etiquetas

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Relacion entre total de la cuenta")
plt.xlabel("Total")
plt.ylabel("SubTotal")
plt.show()
```
