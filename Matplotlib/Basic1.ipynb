Comandos básicos de Matplotlib

```python
import matplotlib.pyplot as plt
```

Crear un gráfico simple (líneas)

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])  # Graficar dos listas: una para el eje X y otra para el eje Y
plt.show()  # Mostrar el grafico
```

Títulos y etiquetas

```python
plt.title("Título del Gráfico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

Cambiar el estilo de la línea y color

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], linestyle='--', color='g')  # Línea discontinua de color rojo
plt.show()
```

Gráficos de dispersión (scatter plot)

```python
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])  # Crear un gráfico de dispersión
plt.show()
```

Gráficos de barras

```python
plt.bar([1, 2, 3], [10, 20, 15])  # Gráfico de barras: X -> [1, 2, 3], Y -> [10, 20, 15]
plt.show()
```

Gráficos de pastel

```python
etiquetas = ['A', 'B', 'C']
valores = [50, 30, 20]
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%')  # Gráfico de pastel con porcentaje
plt.show()
```

Subplots (múltiples gráficos en una figura)

```python
fig, axs = plt.subplots(2, 2)  # Crear una figura con una cuadrícula de 2x2 subplots
axs[0, 0].plot([1, 2, 3], [1, 4, 9])  # Gráfico en la posición [0, 0]
axs[1, 1].bar([1, 2, 3], [10, 20, 30])  # Gráfico de barras en la posición [1, 1]
plt.show()
```

Guardar gráficos como imagen

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.savefig('grafico.png')  # Guardar el gráfico como archivo PNG
plt.show()
```
